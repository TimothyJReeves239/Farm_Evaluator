```python
dist = pgeocode.GeoDistance('US')
dist.query_postal_code('10004', '13487')
```

```python
listodist = []
for x in farm_data["Zip"]:
    listodist.append(dist.query_postal_code('10004', x))
```

```python
farm_data["Dist2City"] = listodist
```

```python
plt.hist(farm_data["Dist2City"])
plt.show()
```

```python
farm_data["A/P"] = farm_data["Price $"]/farm_data["Acres"]
```

```python
farm_data["DistRank"] = farm_data["Dist2City"].rank()
farm_data["A/P_Rank"] = farm_data["A/P"].rank()
farm_data
```

```python
plt.scatter(farm_data["DistRank"], farm_data["Dist2City"])
plt.show()
plt.scatter(farm_data["A/P_Rank"], farm_data["A/P"])
plt.show()
```
